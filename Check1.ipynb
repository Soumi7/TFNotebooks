```
from simpletransformers.ner import NERModel
import pandas as pd
import logging


logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)

# Creating train_df  and eval_df for demonstration
train_data = [[0, 'Consumer', 'B-sector'], [0, 'Discretionary', 'I-sector'],
 [1, 'Creditors', 'B-funda'],
 [2, 'Cos', 'O'], 
 [2, 'where', 'O'], 
 [2, 'ROE', 'B-funda'] ,[2, '>', 'operator'], [2, '15%', 'threshold'],
 [2, 'and', 'Join'],
 [2, 'profit', 'B-funda'],
 [2, 'growth', 'B-attr'],
 [2, '>', 'operator'],
 [2, '20%', 'threshold'],
 [3, 'Tax', 'B-funda'], [3, 'rate', 'I-funda'],
 [4, 'Cos', 'O'], [4, 'where', 'O'],
 [4, '2', 'TPQty'],
 [4, 'year', 'TPUnit'],
 [4, 'revenue', 'B-sector'],
 [4, 'growth', 'B-attr'],
 [4, '>', 'operator']]

train_df = pd.DataFrame(train_data, columns=['sentence_id', 'words', 'labels'])

eval_data = [[0, 'Short', 'B-funda'],
 [0, 'term', 'I-funda'],
 [0, 'debt', 'I-funda'],
 [1, 'Networth', 'B-funda'],
 [2, 'Fastest', 'Sortby'],
 [2, 'revenue', 'B-funda'],
 [2, 'growth', 'B-attr'],
 [2, 'cos', 'O'],
 [3, 'Other', 'B-funda'],
 [3, 'earning', 'I-funda'],
 [3, 'assets', 'I-funda'],
 [4, 'Health', 'B-sector'],
 [4, 'Care', 'I-sector']
 ]
eval_df = pd.DataFrame(eval_data, columns=['sentence_id', 'words', 'labels'])

# Create a NERModel
model = NERModel('bert', 'bert-base-cased', use_cuda=False, args={'overwrite_output_dir': True, 'reprocess_input_data': True})

# Train the model
model.train_model(train_df)

# Evaluate the model
result, model_outputs, predictions = model.eval_model(eval_df)

# Predictions on arbitary text strings
predictions, raw_outputs = model.predict(["Short term debt"])

print(predictions)
```
