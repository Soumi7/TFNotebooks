```
df_train['Date'] = pd.to_datetime(df_train['Date'],format="%d.%m.%Y")
df_train.dtypes
```

```
df_train.replace(['WINE_ALCOHOL 18%', 'BEER','VODKA','VODKA_ALCOHOL'], ['ALCOHOL','ALCOHOL','ALCOHOL','ALCOHOL'], inplace=True)
df_train.Pgroup.unique()
```

```
df_train.columns
```

```
df_train = df_train.rename(columns={
        'Date': 'date',
        'PKod': 'sku',
        'Pgroup': 'group',
        'Pname': 'name',
        'Pquantity': 'quantity',
        'pce_zn': 'unit_cogs',
        'pwa_zn': 'agg_cogs',  
        'pce_sn': 'unit_revenue',
        'pwn_sn': 'agg_rev',  
        'pmarza': 'gross_margin',
        'pmarzajedn': 'unit_contribution',
        'pkwmarza': 'agg_dollar_contribution',
        'pudzmarza': 'share_of_margin'
    })
```

```
df_train.columns
```

```
df_train = df_train.drop(['pwa_sn', 'pce_sb', 'pwa_sb', 'pudzsb'], axis=1)
```

```
df_train['quantity'].describe()
```

```
import seaborn as sns
sns.distplot(df_train['quantity'])
```

```
import matplotlib.pyplot as plt
corrmatrix= df_train.corr()
f,ax = plt.subplots(figsize=(12,9))
sns.heatmap(corrmatrix, vmax=.8, square=True)
```

```
plt.scatter(df_train['quantity'],df_train['share_of_margin'])
plt.title("Corelation between quantity and margin ")
plt.xlabel("Quantity[pieces]")
plt.ylabel("Margin [%]")
plt.show()
```

```
sns.set()
cols=['date', 'sku', 'group', 'name', 'quantity', 'unit_cogs', 'agg_cogs','unit_revenue', 'gross_margin','unit_contribution', 'agg_dollar_contribution', 'share_of_margin']
sns.pairplot(df_train[cols], height=2.5)
plt.show()
```
